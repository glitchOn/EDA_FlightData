        # Notebook Guide
        
        - `flight_eda_notebook.ipynb`: primary notebook that mirrors the scripted phases and includes quick EDA cells.
        - Data lives in `../data/`; plots save to `../plots/`.
        - Use the Julia 1.10 kernel for the best experience.
        

- Sample data: use `../data/flight_data_2024_sample.csv` (copy to `flight_data_2024.csv` if running the scripts end-to-end). Full raw/cleaned files are not in Git due to size.
